```python
import requests
from bs4 import BeautifulSoup
import os

# Global URL to be used across functions
# URL = "https://cointelegraph.com/tags/bitcoin"
URL = "https://cointelegraph.com/tags/bitcoin"

def download_html(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }
    file_name = f"{url.split('/')[-1]}.html"
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an HTTPError for bad requests (4XX or 5XX)
        
        # Save the content to an HTML file
        with open(file_name, 'w', encoding='utf-8') as file:
            file.write(response.text)
        print(f"HTML content saved to {file_name}")
        return file_name
    except requests.RequestException as e:
        print(f"An error occurred: {e}")
        return None

def extract_titles_and_links(file_name):
    with open(file_name, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')

    a_tags = soup.find_all('a')
    articles = []
    
    for tag in a_tags:
        url = tag.get('href')
        title = tag.get_text(strip=True)
        
        if url and title:
            articles.append((title, url))

    output_file = f"{os.path.splitext(file_name)[0]}_links.txt"
    with open(output_file, "w", encoding="utf-8") as file:
        for title, url in articles:
            file.write(f"Title: {title}, URL: {url}\n")
    
    print(f"Titles and URLs have been saved to {output_file}")

# Main execution
if __name__ == "__main__":
    file_name = download_html(URL)
    if file_name:
        extract_titles_and_links(file_name)
```

```python
import requests
from bs4 import BeautifulSoup

# Base URL for concatenation
BASE_URL = "https://cointelegraph.com"

def extract_titles_links_timestamps(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an HTTPError for bad requests (4XX or 5XX)
        soup = BeautifulSoup(response.text, 'html.parser')

        articles = []
        # Assuming each article is contained within an article tag
        for article in soup.find_all('article'):
            title_tag = article.find('a')  # This assumes the title is within the first <a> tag of each article
            timestamp_tag = article.find('time')  # This assumes the timestamp is within a <time> tag
            
            if title_tag and timestamp_tag:
                title = title_tag.get_text(strip=True)
                url = title_tag.get('href')
                full_url = f"{BASE_URL}{url}"  # Create full URL
                timestamp = timestamp_tag.get_text(strip=True)
                articles.append((title, full_url, timestamp))

        # Print results
        for title, full_url, timestamp in articles:
            print(f"Title: {title}, URL: {full_url}, Timestamp: {timestamp}")
            
    except requests.RequestException as e:
        print(f"An error occurred: {e}")

# Main execution
if __name__ == "__main__":
    extract_titles_links_timestamps(BASE_URL + "/tags/bitcoin")
```

```python
def filter_news_links(file_path, output_file):
    # Initialize a list to hold the filtered titles and URLs
    filtered_links = []

    # Read from the provided file
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    # Process each line to find the relevant links
    for line in lines:
        if '/news' in line:
            title = line.split(', URL: ')[0].strip().replace('Title: ', '')
            url = line.split(', URL: ')[1].strip()
            # Prepend the base URL to the extracted URL
            full_url = f"https://cointelegraph.com{url}"
            filtered_links.append((title, full_url))

    # Save the filtered titles and URLs to a new file
    with open(output_file, 'w', encoding='utf-8') as out_file:
        for title, url in filtered_links:
            out_file.write(f"Title: {title}, URL: {url}\n")

    print(f"Filtered links have been saved to {output_file}")

# Specify the path to the input and output files
file_path = r'D:\BlackWhale\WebScaper\bitcoin_links.txt'
output_file = 'filtered_news_links11.txt'

# Call the function
filter_news_links(file_path, output_file)
```

```python
def extract_urls(file_path):
    # Initialize an empty list to store the URLs
    urls = []

    # Read from the provided file
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    # Process each line to extract the URL
    for line in lines:
        # Split each line to extract the URL part
        parts = line.strip().split(', URL: ')
        if len(parts) == 2:
            url = parts[1]  # Extract the URL part
            urls.append(url)

    return urls

# Specify the path to the input file
file_path = r"D:\BlackWhale\WebScaper\filtered_news_links11.txt"

# Call the function to extract URLs from the file
urls_list = extract_urls(file_path)
print(urls_list)

# Print the list of URLs
print("List of URLs:")
for url in urls_list:
    print(url)

```
